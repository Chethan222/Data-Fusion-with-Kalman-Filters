```python
import numpy as np
import matplotlib.pyplot as plt
```

### Generating random data

```python
n = 10
std = 15/np.sqrt(n)
mean = 100
X = (np.random.randn(n)*std-mean).reshape(-1,1)
Y = (2*X + 21).reshape(-1,1) + (np.random.randn(n)*.2-4).reshape(-1,1)

data = np.concatenate((X,Y),axis=1)

plt.plot(X,Y)
```

$\hat{x} = (HH^T)^{-1}H^Ty$

```python
def leastSquaresSol(H,Y):
    H = np.array(H)
    X = np.matmul(np.matmul(np.linalg.inv(np.matmul(H.T,H)),H.T),Y)
    return X.T
  
```

```python
def lineOfBestFit(data):
    ones = np.ones((data.shape[0],1))
    H = np.concatenate((data[:,0].reshape(-1,1),ones),axis=1)
    Y = data[:,1]
    lineParams = leastSquaresSol(H, Y)
    return lineParams
```

```python
params = lineOfBestFit(data)
x = list(range(int(data[:,0].min()),int(data[:,0].max())))
y = [params[1]+params[0]*i for i in x ]
plt.plot(x,y)
```
